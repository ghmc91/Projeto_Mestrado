```python
df_reduced = df[['V005', 'ANTENNA_ID']]
df_reduced = df_reduced.groupby('ANTENNA_ID').mean().reset_index()
```

```python
#Criando a coluna classe para classificar os setores sensitários de acordo com sua renda
def classe(i):
    if i <= 510:
        return 1
    elif i <= 1020:
        return 2
    elif i <= 1530:
        return 3
    elif i <= 2550:
        return 4
    elif i <= 5100:
        return 5
    elif i <= 10200
        return 6
    elif i > 10200:
        return 7

df_reduced['classe'] = df_reduced['V005'].map(classe)
df_reduced = df_reduced.sort_values(['classe'])
```
