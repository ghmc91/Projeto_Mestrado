# Tratando Valores missing

```python
def missing_values_table(df):
        mis_val = df.isnull().sum()
        
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        print ("Seu dataframe tem " + str(df.shape[1]) + " colunas.\n"      
            "Há " + str(mis_val_table_ren_columns.shape[0]) +
              " colunas que possuem valores ausentes.")
        
        return mis_val_table_ren_columns
```

```python
missing_values_table(sjdr_data)
```

```python
sjdr_data = sjdr_data.drop('Unnamed: 33', axis=1)
```

```python
#Conversão do tipo de valor da V005
sjdr_data['V005'] = sjdr_data.V005.apply(str)
sjdr_data['V005'] = sjdr_data['V005'].apply(lambda x: str(x.replace(',', '.')))
sjdr_data['V005'] = sjdr_data.V005.replace('NaN', np.nan)
sjdr_data['V005'] = sjdr_data.V005.apply(float)
```

```python
sjdr_data.sample(5)
```

```python
#Criando a coluna classe para classificar os setores sensitários de acordo com sua renda
def classe(i):
    if i <= 510:
        return 1
    elif i <= 1020:
        return 2
    elif i <= 1530:
        return 3
    elif i <= 2550:
        return 4
    elif i <= 5100:
        return 5
    elif i <= 10200:
        return 6
    elif i > 10200:
        return 7

sjdr_data['classe'] = sjdr_data['V005'].map(classe)
sjdr_data = sjdr_data.sort_values(['classe'])
```

```python
#Contabilizando os setores censitários por classe
sjdr_data['classe'].value_counts(dropna=True)
```

```python
#Conversão para csv
sjdr_data.to_csv('/home/gustavo/Desktop/Mestrado/mestrado_dados/ibge/sjdr_data.csv')
```

```python
#Agrupando os valores por subdistrito e calculando a média do ganho
#dos setores censitários que constam no subdistrito
cities_data = pd.DataFrame(sjdr_data.groupby('Nome_do_municipio')['V005'].mean())
```

```python
#atribuindo a classe de acordo com rendimento
cities_data['classe'] = cities_data['V005'].map(classe)
```
