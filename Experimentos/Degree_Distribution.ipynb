```python
def segregate_network(path, isDirected):
    G = nx.read_gml(path)
    if isDirected:
        cl = nx.get_node_attributes(G, 'class')
        G = G.subgraph(list(cl))        
    G_1 = [x for x,y in G.nodes(data=True) if y['class'] == 1]
    G_2 = [x for x,y in G.nodes(data=True) if y['class'] == 2]
    G_3 = [x for x,y in G.nodes(data=True) if y['class'] == 3]
    G_4 = [x for x,y in G.nodes(data=True) if y['class'] == 4]
    G_5 = [x for x,y in G.nodes(data=True) if y['class'] == 5]
    G_6 = [x for x,y in G.nodes(data=True) if y['class'] == 6]
    G_7 = [x for x,y in G.nodes(data=True) if y['class'] == 7]
    list_G = [G_1, G_2, G_3, G_4, G_5, G_6, G_7]
    list_G_not_empty = []
    for g in list_G:
        if g:
            list_G_not_empty.append(g)
    subgraphs = []
    for n in list_G_not_empty:
        subgraphs.append(G.subgraph(n))
    return subgraphs
```
