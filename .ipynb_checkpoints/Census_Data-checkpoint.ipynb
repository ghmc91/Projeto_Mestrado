# Tratando Valores missing

```python
def missing_values_table(df):
        mis_val = df.isnull().sum()
        
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        print ("Seu dataframe tem " + str(df.shape[1]) + " colunas.\n"      
            "Há " + str(mis_val_table_ren_columns.shape[0]) +
              " colunas que possuem valores ausentes.")
        
        return mis_val_table_ren_columns
```

```python
missing_values_table(dados_rj)
```

```python
dados_rj=dados_rj.drop('Unnamed: 33', axis=1)
```

```python
#Conversão do tipo de valor da V005
dados_rj['V005'] = dados_rj.V005.apply(str)
dados_rj['V005'] = dados_rj['V005'].apply(lambda x: str(x.replace(',', '.')))
dados_rj['V005'] = dados_rj.V005.replace('NaN', np.nan)
dados_rj['V005'] = dados_rj.V005.apply(float)
```

```python
dados_rj.sample(5)
```

```python
#Criando a coluna classe para classificar os setores sensitários de acordo com sua renda
def classe(i):
    if i < 510:
        return 1
    elif i < 1020:
        return 2
    elif i < 1530:
        return 3
    elif i < 2550:
        return 4
    elif i < 5100:
        return 5
    elif i < 10200:
        return 6
    elif i > 10200:
        return 7

dados_rj['classe'] = dados_rj['V005'].map(classe)
dados_rj = dados_rj.sort_values(['classe'])
```

```python
#Contabilizando os setores censitários por classe
dados_rj['classe'].value_counts(dropna=True)
```

```python
#Conversão para csv
dados_rj.to_csv('dados_rj.csv')
```

```python
missing_values_table(dados_rj)
```

```python
#Agrupando os valores por subdistrito e calculando a média do ganho
#dos setores censitários que constam no subdistrito
dados_subdist = pd.DataFrame(dados_rj.groupby('Nome_do_subdistrito')['V005'].mean())
```

```python
#atribuindo a classe de acordo com rendimento
dados_subdist['classe'] = dados_subdist['V005'].map(classe)
```

```python
dados_subdist.to_csv('dados_subdist.csv')
```
