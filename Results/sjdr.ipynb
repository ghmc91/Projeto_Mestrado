```python
# dict_users_num_calls_sjdr => Chave: Usuário/Valor: Número de ligações por usuário
file = open('/home/gustavo/Desktop/Mestrado/mestrado_dados/ProcessamentoVinicius/Sjdr/dict_users_num_calls_sjdr.pickle', 'rb')
dict_users_num_calls_sjdr = pickle.load(file)
dict_users_num_calls_sjdr
```

```python
# dict_users_residence_antenna_sjdr => Chave: Usuário
#Valor: Antena mais próxima da região presumida como de residência do usuário
file = open('/home/gustavo/Desktop/Mestrado/mestrado_dados/ProcessamentoVinicius/Sjdr/dict_users_residence_antenna_sjdr.pickle', 'rb')
dict_users_residence_antenna_sjdr = pickle.load(file)
dict_users_residence_antenna_sjdr
```

```python
for key, value in zip(dict_users_residence_antenna_sjdr.keys(), dict_users_residence_antenna_sjdr.values()):
    dict_users_residence_antenna_sjdr[key] = int(value)
```

```python
#Dicionário contendo a residencia e a classificação econômica
df_unique_antennas = pd.read_csv('/home/gustavo/Desktop/Mestrado/mestrado_dados/ProcessamentoVinicius/Sjdr/antennas_sjdr_unique.txt', sep=';')
df_unique_antennas['ECONOMIC_CLASSIFICATION'] = [3 if (x == 'São João del-Rei') | (x == 'Tiradentes') else 2 for x in df_unique_antennas['CITY']]
df_cellid_econ_class = df_unique_antennas[['CELLID', 'ECONOMIC_CLASSIFICATION']]
df_cellid_econ_class = df_cellid_econ_class.set_index('CELLID')
dict_econ_class = df_cellid_econ_class.to_dict('index')
dict_merge = {}
for key_1, value_1 in zip(dict_users_residence_antenna_sjdr.keys(), dict_users_residence_antenna_sjdr.values()):
    for key_2, value_2 in zip(dict_econ_class.keys(), dict_econ_class.values()):
        if value_1 == key_2:
            dict_merge[key_1] = {'ECON_CLASS': list(value_2.values())[0], 'RESIDENCE': value_1}
dict_merge
```

```python
# dict_users_stays_weekdays_sjdr => Chave: Usuário/Valor: Locais de ligações do usuário nos dias de semana
# Adicionando a esse dicionário o horário das ligações é possível criar a matriz de localização
# e os vetores de trajetória
file = open('/home/gustavo/Desktop/Mestrado/mestrado_dados/ProcessamentoVinicius/Sjdr/dict_users_stays_weekdays_sjdr.pickle', 'rb')
dict_users_stays_weekdays_sjdr = pickle.load(file)
dict_users_stays_weekdays_sjdr
```

```python
# dict_users_stays_weekends_sjdr => Chave: Usuário/Valor: Locais de ligações do usuário nos fins de semana
file = open('/home/gustavo/Desktop/Mestrado/mestrado_dados/ProcessamentoVinicius/Sjdr/dict_users_stays_weekends_sjdr.pickle', 'rb')
dict_users_stays_weekends_sjdr = pickle.load(file)
dict_users_stays_weekends_sjdr
```

```python
#Esse dicionário (dict_residence_vl_weekdays) contêm os locais de residência e os locais de visitação dos usuários nos dias de semana
intersection_weekdays = dict_users_residence_antenna_sjdr.keys() & dict_users_stays_weekdays_sjdr.keys()
dict_residence_vl_weekdays = {}
for i in intersection_weekdays:
    residence = dict_users_residence_antenna_sjdr.get(i)
    locals_weekday = dict_users_stays_weekdays_sjdr.get(i)
    new_dict = {i: {'Residence': residence, 'Visited_Locals': locals_weekday}}
    dict_residence_vl_weekdays.update(new_dict)
dict_residence_vl_weekdays
```

```python
intersection_weekend = dict_users_residence_antenna_sjdr.keys() & dict_users_stays_weekends_sjdr.keys()
dict_residence_vl_weekend = {}
for i in intersection_weekend:
    residence = dict_users_residence_antenna_sjdr.get(i)
    locals_weekend = dict_users_stays_weekends_sjdr.get(i)
    new_dict = {i: {'Residence': residence, 'Visited_Locals': locals_weekend}}
    dict_residence_vl_weekend.update(new_dict)
dict_residence_vl_weekend
```
