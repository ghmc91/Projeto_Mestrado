```python
for key, value dict_users_residence_antenna_jf.items():
    dict_users_residence_antenna_jf[key] = int(value)
```

```python
#Dicionrio contendo a residencia e a classificao econmica
df_unique_antennas = pd.read_csv('/home/gustavo/Desktop/Mestrado/mestrado_dados/ProcessamentoVinicius/Jf/antennas_jf_unique.txt', sep=';')
df_unique_antennas['ECONOMIC_CLASSIFICATION'] = [4 if (x == 'Juiz de Fora') 
                                                 else 3 if (x == 'Chiador') | (x == 'Matias Barbosa')
                                                 else 2
                                                 for x in df_unique_antennas['CITY']]
df_cellid_econ_class = df_unique_antennas[['CELLID', 'ECONOMIC_CLASSIFICATION']]
df_cellid_econ_class = df_cellid_econ_class.set_index('CELLID')
dict_econ_class = df_cellid_econ_class.to_dict('index')
dict_merge = {}
for key_1, value_1 in dict_users_residence_antenna_jf.items():
    for key_2, value_2 in dict_econ_class.items():
        if value_1 == key_2:
            dict_merge[key_1] = {'ECON_CLASS': list(value_2.values())[0], 'RESIDENCE': value_1}
dict_merge
```

```python
file = open('/home/gustavo/Desktop/Mestrado/mestrado_dados/ProcessamentoVinicius/Jf/dict_users_stays_weekdays_jf.pickle', 'rb')
dict_users_stays_weekdays_jf = pickle.load(file)
dict_users_stays_weekdays_jf
```

```python
intersection_weekdays = dict_users_residence_antenna_jf.keys() & dict_users_stays_weekdays_jf.keys()
dict_residence_vl_weekdays = {}
for i in intersection_weekdays:
    residence = dict_users_residence_antenna_jf.get(i)
    locals_weekday = dict_users_stays_weekdays_jf.get(i)
    new_dict = {i: {'Residence': residence, 'Visited_Locals': locals_weekday}}
    dict_residence_vl_weekdays.update(new_dict)
dict_residence_vl_weekdays
```
